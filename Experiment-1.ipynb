```python
# Training data: (Attributes, Class Label)
# Example: EnjoySport dataset

training_data = [
    (["Sunny", "Warm", "Normal", "Strong", "Warm", "Same"], "Yes"),
    (["Sunny", "Warm", "High",   "Strong", "Warm", "Same"], "Yes"),
    (["Rainy", "Cold", "High",   "Strong", "Warm", "Change"], "No"),
    (["Sunny", "Warm", "High",   "Strong", "Cool", "Change"], "Yes")
]
```

```python
# Initialize hypothesis with most specific values (Ï•)
hypothesis = ["0"] * len(training_data[0][0])

print("Initial Hypothesis:")
print(hypothesis)
```

```python
for instance, label in training_data:
    if label == "Yes":  # Only consider positive examples
        for i in range(len(hypothesis)):
            if hypothesis[i] == "0":
                hypothesis[i] = instance[i]
            elif hypothesis[i] != instance[i]:
                hypothesis[i] = "?"

    print("\nUpdated Hypothesis:")
    print(hypothesis)
```

```python
print("\nFinal Most Specific Hypothesis found by FIND-S:")
print(hypothesis)
```
