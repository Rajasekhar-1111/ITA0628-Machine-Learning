```python
import pandas as pd

data = pd.read_csv("training_data.csv")
concepts = data.iloc[:, :-1].values
labels = data.iloc[:, -1].values

print(concepts)
print(labels)
```

```python
# Initialize Specific and General hypotheses
num_attributes = concepts.shape[1]

S = ["0"] * num_attributes
G = [["?"] * num_attributes]

print("Initial S:", S)
print("Initial G:", G)
```

```python
for i, example in enumerate(concepts):
    if labels[i] == "Yes":
        # Positive example
        for j in range(num_attributes):
            if S[j] == "0":
                S[j] = example[j]
            elif S[j] != example[j]:
                S[j] = "?"
        G = [g for g in G if all(g[k] == "?" or g[k] == S[k] for k in range(num_attributes))]

    else:
        # Negative example
        new_G = []
        for g in G:
            for j in range(num_attributes):
                if g[j] == "?" and S[j] != example[j]:
                    new_hypothesis = g.copy()
                    new_hypothesis[j] = S[j]
                    new_G.append(new_hypothesis)
        G = new_G

    print(f"\nAfter example {i+1}:")
    print("S:", S)
    print("G:", G)
```

```python
print("\nFinal Specific Hypothesis (S):", S)
print("Final General Hypotheses (G):", G)
```
